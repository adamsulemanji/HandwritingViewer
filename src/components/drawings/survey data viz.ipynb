```python
import numpy as np
import matplotlib.pyplot as plt

# Data
data = [22, 24, 30, 22, 16, 1]

# Calculate statistics
mean_value = np.mean(data)
median_value = np.median(data)
first_quartile = np.percentile(data, 25)
third_quartile = np.percentile(data, 75)
iqr = third_quartile - first_quartile

# Print the statistical summary
print("Statistical Summary:")
print("Mean:", mean_value)
print("Median:", median_value)
print("First Quartile (Q1):", first_quartile)
print("Third Quartile (Q3):", third_quartile)
print("Interquartile Range (IQR):", iqr)

# Create a figure with two subplots (histogram and box plot)
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 6))

# Plot histogram
ax1.hist(data, bins=10, edgecolor='black', alpha=0.7)
ax1.set_title('Years Worked in Education')
ax1.set_xlabel('Years')
ax1.set_ylabel('Frequency')

# Plot box plot
ax2.boxplot(data, vert=False)
ax2.set_title('Years Worked in Education')
ax2.set_xlabel('Years')

# Show statistical summary on the box plot at the top right corner
summary_text = f"Mean: {mean_value:.2f}\nMedian: {median_value}\nQ1: {first_quartile}\nQ3: {third_quartile}\nIQR: {iqr}"
ax2.text(0.95, 0.95, summary_text, verticalalignment='top', horizontalalignment='right', transform=ax2.transAxes, fontsize=10, bbox=dict(facecolor='white', alpha=0.5))

# Adjust layout and display the plot
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Your data in a dictionary format
data = {
    'Ease of use': [6, 4, 4, 6, 6, 4, 4],
    'The tool is easy to learn and understand': [6, 6, 4, 6, 6, 5, 6],
    'The application is aesthetically pleasing': [6, 6, 4, 3, 6, 5, 5],
    'The tool is quick to use': [6, 6, 4, 6, 6, 5, 6],
    'Ability to compare students work to others': [6, 6, 1, 6, 6, 6, 4],
    'Ability to see a student\'s writing over time': [6, 6, 6, 6, 6, 6, 6],
    'Ability to see details of student\'s handwriting': [6, 6, 4, 6, 6, 6, 6],
    'Ability to leave notes about student\'s writing': [6, 6, 1, 3, 6, 5, 6],
    'Ability to leave ratings about student\'s writing': [6, 6, 1, 3, 6, 5, 6]
}

# Create a DataFrame from the data
df = pd.DataFrame(data)

# Calculate the mean values for each column
mean_values = df.mean()

# Plotting the mean Likert scores
plt.figure(figsize=(10, 6))
mean_values.sort_values(ascending=False).plot(kind='bar', color='skyblue')
plt.title('Mean Likert Scale Scores')
plt.xlabel('Attributes')
plt.ylabel('Mean Score')
plt.xticks(rotation=45, ha='right')
plt.ylim(0, 7)  # Set y-axis range from 0 to 7 (assuming Likert scale is from 1 to 6)
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Your data in a dictionary format
data = {
    'Ease of use': [6, 4, 4, 6, 6, 4, 4],
    'The tool is easy to learn and understand': [6, 6, 4, 6, 6, 5, 6],
    'The application is aesthetically pleasing': [6, 6, 4, 3, 6, 5, 5],
    'The tool is quick to use': [6, 6, 4, 6, 6, 5, 6],
    'Ability to compare students work to others': [6, 6, 1, 6, 6, 6, 4],
    'Ability to see a student\'s writing over time': [6, 6, 6, 6, 6, 6, 6],
    'Ability to see details of student\'s handwriting': [6, 6, 4, 6, 6, 6, 6],
    'Ability to leave notes about student\'s writing': [6, 6, 1, 3, 6, 5, 6],
    'Ability to leave ratings about student\'s writing': [6, 6, 1, 3, 6, 5, 6]
}

# Create a DataFrame from the data
df = pd.DataFrame(data)

# Set up the plotting environment
sns.set(style="whitegrid", font_scale=1.2)

# Plot each column separately
for column in df.columns:
    plt.figure(figsize=(8, 5))
    sns.histplot(df[column], bins=range(1, 8), discrete=True, kde=False, color='skyblue', edgecolor='black')
    plt.title(f'Distribution of Responses - {column}')
    plt.xlabel('Likert Scale Value')
    plt.ylabel('Count')
    plt.xticks(ticks=range(1, 7), labels=[str(i) for i in range(1, 7)])
    plt.show()
```
